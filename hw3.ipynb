```python
print("Задача 4.1")
print('Дано:')
display(Math(r"""A_1 x + B_1 y+ C_1 z + D_1 = 0"""))
print("Для того чтобы записать уравнение плоскости параллельной данной необходимо и достаточно чтобы система уравнений вида:")
display(Math(r"""
\begin{cases}
    A_1 x + B_1 y+ C_1 z + D_1 = 0\\
    A_2 x + B_2 y+ C_2 z + D_2 = 0
\end{cases}"""))
print("""не имела решений.\nЕсли такая система не имеет решений, \
то не существует ни одной точки в прямоугольной системе \
координат Oxyz в трехмерном пространстве, координаты которой \
удовлетворяют одновременно обоим уравнениям системы.""")
print("Рассмотрим две матрицы составленные из системы:")
display(Math(r""" M = 
 \begin{pmatrix}
  A_1 & B_1 & C_1 \\
  A_2 & B_2 & C_2 \\ 
 \end{pmatrix}"""))
print("и расширенную:")
display(Math(r""" \widetilde{M} = 
 \begin{pmatrix}
  A_1 & B_1 & C_1 & D_1\\
  A_2 & B_2 & C_2 & D_2\\ 
 \end{pmatrix}"""))
print("Если все определители основной матрицы будут \
равны нулю и следовательно ранг основной матрицы будет равен 1. \
А определитель расширенной матрицы, к примеру:")
display(Math(r"""
 \begin{vmatrix}
  A_1 & D_1\\
  A_2 & D_2\\ 
 \end{vmatrix} = A_1 D_2 - A_2 D_1 \neq 0\\
 A_1 D_2 \neq A_2 D_1"""))
print("следовательно ранг расширенной равен двум. \
Если ранг основной матрицы системы уравнений меньше \
ранга расширенной матрицы системы, то согласно теореме Кронекера-Капелли \
следует, что система уравнений решений не имеет, \
тогда плоскости будут параллельны.")
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,1)
Y = np.arange(-5,5,1)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 3*Y
Z2 = 2*X + 3*Y +50
ax.plot_surface(X,Y,Z)
ax.plot_surface(X,Y,Z2)
ax.scatter(0,0,0,'z',50,'red')
show()
```
